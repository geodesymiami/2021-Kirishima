## Supplementary Figure S2-S4: displacement time-series map

+ Figure S2 - AlosAT424
+ Figure S3 - AlosDT73
+ Figure S4 - Alos2DT23

```python
%matplotlib inline
import os
from mintpy import view
```

### ALOS-1 asc 424

```python
proj_dir = os.path.expanduser('~/insarlab/Kirishima/ShinmoeAlosAT424/mintpy')
ts_file = os.path.join(proj_dir, 'timeseries_ERA5_ramp_demErr.h5')
dem_file = os.path.join(proj_dir, 'inputs/geometryGeo.h5')

out_fig = os.path.join(proj_dir, 'dis_ts_map_AlosAT424.png')

scp_args = '{t} --nrows 6 --ncols 5 -v -5 5 --figsize 10 12.5 --dem {d}'.format(t=ts_file, d=dem_file)
scp_args += ' --wspace 0.1 --hspace 0.1 '
scp_args += ' --dpi 300 -o {o} '.format(o=out_fig) # output config
view.main(scp_args.split())
```

### ALOS-1 desc 73

```python
proj_dir = os.path.expanduser('~/insarlab/Kirishima/ShinmoeAlosDT73/mintpy')
ts_file = os.path.join(proj_dir, 'timeseries_ERA5_ramp_demErr.h5')
dem_file = os.path.join(proj_dir, 'inputs/geometryGeo.h5')

out_fig = os.path.join(proj_dir, 'dis_ts_map_AlosDT73.png')

scp_args = '{t} --nrows 4 --ncols 5 -v -5 5 --figsize 9 7.5 --dem {d} '.format(t=ts_file, d=dem_file)
scp_args += ' --wspace 0.1 --hspace 0.05 '
scp_args += ' --dpi 300 -o {o} '.format(o=out_fig) # output config
view.main(scp_args.split())
```
