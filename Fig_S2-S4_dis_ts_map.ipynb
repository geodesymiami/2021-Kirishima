```python
%matplotlib inline

import os
from mintpy import view

work_dir = os.path.expanduser('~/Documents/Paper/2019_Kirishima/notebooks/Shinmoe_data')
os.chdir(work_dir)
print('Go to directory', work_dir)

# common output figure setting
opt = ' --dpi 300 --save '
```

```python
## AlosDT73
proj_dir = os.path.expanduser('~/insarlab/Kirishima/ShinmoeAlosDT73/mintpy')
ts_file = os.path.join(proj_dir, 'timeseries_ERA5_ramp_demErr.h5')
dem_file = os.path.join(proj_dir, 'inputs/geometryGeo.h5')

out_fig = os.path.join(proj_dir, 'dis_ts_map_AlosDT73.png')

scp_args = '{t} --nrows 4 --ncols 5 -v -6 6 --figsize 10 8 --dem {d} '.format(t=ts_file, d=dem_file)
scp_args += ' {p} -o {o} '.format(p=opt, o=out_fig) # output config
view.main(scp_args.split())
```
