# Prepare .mat files for modeling with GBIS

```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
from mintpy.utils import readfile
from mintpy import save_gbis


## Common setting
out_dir = os.path.expanduser('~/insarlab/Kirishima/Model/data')

## Dataset Initiation
proj_dir_base = os.path.expanduser('~/insarlab/Kirishima')
vel_files = [os.path.join(proj_dir_base, i) for i in [
    'KirishimaAlosAT424/mintpy/velocity.h5',
    'KirishimaAlosDT73/mintpy/velocity.h5',
    'KirishimaAlos2AT131/mintpy/velocity.h5',
    'KirishimaAlos2DT23/mintpy/velocity.h5',
    'KirishimaAlos2AT131/mintpyAll/velocity2017post.h5',
    'KirishimaAlos2DT23/mintpyAll/velocity2017post.h5',
]]

for vel_file in vel_files:
    # grab project and file names
    proj_dir = os.path.dirname(vel_file)
    date12 = readfile.read_attribute(vel_file)['DATE12']
    unw_file = os.path.join(proj_dir, '{}.unw'.format(date12))
    geom_file = os.path.join(proj_dir, 'inputs/geometryGeo.h5')

    # run
    scp_args = '{} -g {} --out-dir {} --ellipsoid2geoid '.format(unw_file, geom_file, out_dir)
    # mask out the crater for the Shinmoe-dake 2017 pre-eruptive signal
    if 'Alos2' in vel_file:
        mask_aoi_file = os.path.join(out_dir, 'maskShinmoeCrater2017pre.h5')
        scp_args += ' --mask {} '.format(mask_aoi_file)
    mat_file = save_gbis.main(scp_args.split())
print('Done.')
```
