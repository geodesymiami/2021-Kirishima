# Post-processing of `smallbaselineApp.py`

to extract displacement for the time period of interest

1. Convert average velocity to displacement using `save_roipacl.py`
2. Decompose asc / desc LOS displacement to horz / vert direction using `asc_desc2horz_vert.py`

```python
%matplotlib inline
import os
import numpy as np
from mintpy.utils import readfile, ptime, utils as ut
from mintpy import asc_desc2horz_vert, save_roipac, view, generate_mask

work_dir = os.path.expanduser('~/Papers/2021_Kirishima/figs_src/obs')
os.chdir(work_dir)
print('Go to directory:', work_dir)

## Dataset Initiation
proj_dir_base = os.path.expanduser('~/data/Kirishima')
vel_files = [os.path.join(proj_dir_base, i) for i in [
    'KirishimaAlosAT424/mintpy/velocity.h5',
    'KirishimaAlosDT73/mintpy/velocity.h5',
    'KirishimaAlos2AT131/mintpy/velocity.h5',
    'KirishimaAlos2DT23/mintpy/velocity.h5',
    'KirishimaAlos2AT131/mintpyAll/velocityPost2017.h5',
    'KirishimaAlos2DT23/mintpyAll/velocityPost2017.h5',
]]

msk_files = []
dem_files = []
unw_files = []
date12s = []
for vel_file in vel_files:
    proj_dir = os.path.dirname(vel_file)
    msk_file = os.path.join(proj_dir, 'maskTempCoh.h5')
    dem_file = os.path.join(proj_dir, 'inputs/gsi10m.dem.wgs84')

    # get displacement file name
    date12 = readfile.read_attribute(vel_file)['DATE12']
    unw_file = os.path.join(proj_dir, '{}.unw'.format(date12))

    msk_files.append(msk_file)
    dem_files.append(dem_file)
    unw_files.append(unw_file)
    date12s.append(date12)

## horz / vert files
model_dir = os.path.expanduser('~/Papers/2021_Kirishima/figs_src/data')
models = ['Kirishima2008post',
          'Kirishima2017pre',
          'Kirishima2017post']
hv_files = [os.path.join(model_dir, '{}.h5'.format(model)) for model in models]
```

## 1. Velocity to displacement

```python
# change reference point to a less noisy location for the time period of interest.
# the original reference point is good for the whole time period
# ref_lat, ref_lon = 31.916, 130.850 #31.930, 130.864
ref_lalos = [[31.916, 130.850],
             [31.916, 130.850],
             [31.921, 130.844]]

for i in range(len(vel_files)):
    unw_file = unw_files[i]
    ref_lalo = ref_lalos[int(i/2)]
    tdiff = np.diff(ptime.yyyymmdd2years(date12s[i].split('_')))[0]
    print('time span of displacement: {:.1f} years'.format(tdiff))

    # save to displacement
    scp_args = '{v} -o {o} -m {m} --keep-all-metadata '.format(v=vel_files[i], o=unw_file, m=msk_files[i])
    scp_args += ' --ref-lalo {lat} {lon} '.format(lat=ref_lalo[0], lon=ref_lalo[1])
    save_roipac.main(scp_args.split())

    # plot
    scp_args = '{} phase --wrap --wrap-range -6 6 -u cm --dem {} --noverbose --save'.format(unw_file, dem_files[i])
    view.main(scp_args.split())
```

## 2. Asc / desc to horz / vert

```python
for i in range(len(models)):
    # run
    scp_args = '{f1} {f2} --oo {o} '.format(f1=unw_files[2*i], f2=unw_files[2*i+1], o=hv_files[i])
    print('asc_desc2horz_vert.py', scp_args)
    asc_desc2horz_vert.main(scp_args.split())

    # plot
    opt = '--nrows 2 --ncols 2 --figsize 9 8 -u cm --wrap --wrap-range -6 6 -c cmy --noverbose '
    view.main('{f} {o}'.format(f=hv_files[i], o=opt).split())
```

## 4. Preliminary source geometry based on the ratio of max horz / vert displacement

Typical ratio of max horz/vert displacement (Fig. 4 and 6 in Fialko et al., 2001):
+ Sills: < 0.3
+ Spherical sources: ~0.4

Thus, we model this displacement as **Prolate Spheroid** (Yang, 1988) or **Compound Dislocation Models** (Nikkhoo, 2016).

### Shinmoe-dake 2008 post-eruptive

```python
geo_box = (130.85, 31.94, 130.91, 31.88)   #(W, N, E, S)
hv_file = hv_files[0]

# read data
atr = readfile.read_attribute(hv_files[0])
phase2range = float(atr['WAVELENGTH']) / (-4 * np.pi)
coord = ut.coordinate(atr)
pix_box = coord.box_geo2pixel(geo_box)
hz_data = readfile.read(hv_file, datasetName='horizontal', box=pix_box)[0] * phase2range
up_data = readfile.read(hv_file, datasetName='vertical',   box=pix_box)[0] * phase2range

max_west_dis = -np.nanmin(hz_data)
max_east_dis = np.nanmax(hz_data)
max_vert_dis = np.nanmax(np.abs(up_data))
print('max vert displacement: {:.2f} cm'.format(max_vert_dis))
print('max east displacement: {:.2f} cm; horz/vert ratio: {:.2f}'.format(max_east_dis, max_east_dis/max_vert_dis))
print('max west displacement: {:.2f} cm; horz/vert ratio: {:.2f}'.format(max_west_dis, max_west_dis/max_vert_dis))
```
